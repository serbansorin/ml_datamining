# Construirea modelului

## Selectam din colectia data atributele care ne intereseaza in procesul de modelare (clasificare)

```python
atribute_intrare = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country']
```

```python
X = df[atribute_intrare]
X
```

```python
# Stabilim atributul tinta
y =  df.income
y
```

## Importam pachetele care ne trebuie

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier

import matplotlib.pyplot as plt

# Import datasets, classifiers and performance metrics
from sklearn import datasets, svm, metrics
```

## Transformarea atributelor discrete 

```python
# Selectarea atributelor discrete (de tip Object)

nume_atribute_discrete =  [nume for nume in df.columns if df[nume].dtype == 'object']
nume_atribute_discrete.remove('income')
nume_atribute_discrete
```

```python
# Selectarea atributelor continui (de tip Object)
nume_atribute_continui = [nume for nume in df.columns if df[nume].dtype != 'object']
nume_atribute_continui 
```

```python
X = df[atribute_intrare]

X_discrete = df[nume_atribute_discrete]
# X_continui = df[nume_atribute_continui]

X_copy = X.copy()


#atribute_discrete =  discrete for discrete in df.columns
enc = OrdinalEncoder()
enc.fit(X[nume_atribute_discrete])
X_copy[nume_atribute_discrete] = enc.transform(X[nume_atribute_discrete])
X = X_copy
X
df

#X.dtype
```

# naive-bayes
https://scikit-learn.org/stable/modules/naive_bayes.html  

## Impartim colectia data in multime de training si multime de testare

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
 
```

## Alegerea modelului

```python
# model = GaussianNB()
# model = MultinomialNB()
model = DecisionTreeClassifier(random_state=0)
```

## Aplicarea modelului pe multimea de training

```python
model.fit(X_train, y_train)
```

## Testarea  modelului pe multimea de testare

```python
y_pred = model.predict(X_test)
```

## Evaluarea modelului 

```python
print('Numarul de exemple clasificate incorect este', (y_test != y_pred).sum(), 'din totalul de', X_test.shape[0])
```

## Calcul metrici de evaluare a modelului: 
### eroare, acuratete

```python
error = (y_test != y_pred).sum()/X_test.shape[0]
error
```

```python
acuratetea = 1-error
acuratetea
```

## Matricea de confuzie

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
disp = metrics.ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.title("Matricea de confuzie")
plt.show()
```

## Testarea modelului pe multimea de training

```python
y_pred_train = model.predict(X_train)
```

```python
(y_pred_train != y_train).sum()
error_train = (y_train != y_pred_train).sum()/X_train.shape[0]
error_train
```
